# EXPLORATORY DATA ANALYSIS

```python
import os
import cv2
import matplotlib.pyplot as plt
from PIL import Image
```

```python
def get_filenames(folder_path, start_index, end_index):
    filenames = []
    for i in range(start_index, end_index + 1):
        filename = f"H_{i}.jpg"
        file_path = os.path.join(folder_path, filename)
        if os.path.isfile(file_path):
            filenames.append(filename)
    return filenames

# Provide the path to the folder
folder_path = "/Users/damodargupta/Desktop/breast cancer prediction ann/Breast Cancer Patients MRI's/train/Healthy"

# Set the starting and ending indices
start_index = 1
end_index = 800

# Call the function to get the filenames
filenames = get_filenames(folder_path, start_index, end_index)

print(len(filenames))
    
```
