```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

# Load the image and convert it to grayscale
image_path = "/Users/damodargupta/Desktop/BATCH_19 BREAST-CANCER-DETECTION-ANN/Breast Cancer Patients MRI's/train/Healthy/H_2.jpg"
image = Image.open(image_path).convert('L')

# Convert the grayscale image to a numpy array
image_array = np.array(image)

# Perform 2D Fourier Transform
fft_result = np.fft.fft2(image_array)

# Shift the zero-frequency component to the center
fft_shifted = np.fft.fftshift(fft_result)

# Calculate the magnitude spectrum (frequency information)
magnitude_spectrum = np.abs(fft_shifted)

# Plot the magnitude spectrum
plt.imshow(np.log(1 + magnitude_spectrum), cmap='gray')
plt.title('Magnitude Spectrum (Frequency Information)')
plt.axis('off')
plt.show()
```
