# TESTING THE MODEL

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Load the saved model
model = load_model('/Users/damodargupta/Desktop/breast cancer prediction ann/final-model.h5')
```

```python
# Set the path to the test image
image_path = "/Users/damodargupta/Desktop/breast cancer prediction ann/Breast Cancer Patients MRI's/validation/Healthy/S_32.jpg"

# Set the image dimensions
img_width, img_height = 150, 150

# Load and preprocess the image
img = image.load_img(image_path, target_size=(img_width, img_height))
img_array = image.img_to_array(img)
plt.imshow(img_array)
img_array = np.expand_dims(img_array, axis=0)
img_array /= 255.0

```

```python
# Make predictions
predictions = model.predict(img_array)
print(predictions)
predicted_label = np.round(predictions).flatten()[0]
print(predicted_label)

# Interpret the predictions
if predicted_label == 1:
    result = 'Sick'
else:
    result = 'Healthy'

# Print the predicted label
print(f"Predicted Label: {result}")
```
